```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:/backup/AI course work/Guvi/projects/PatrolIQ Smart Safety Analytics Platform/Data/Crimes_-_2001_to_Present_20251110.csv')
df.shape
```

```python
df = df.sample(n=500000, random_state=42)
df.head()
```

```python
df.columns
```

```python
df.drop(columns = [ 'Case Number', 'Block', 'IUCR', 'Updated On', 'Location'], inplace=True)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Arrest'] = df['Arrest'].astype(int)
df['Domestic'] = df['Domestic'].astype(int)
```

```python
df.value_counts()
```

```python
df.isnull().sum()
```

```python
df['Location Description'].fillna('UNKNOWN', inplace=True)
```

```python
ward_mode = df.groupby('District')['Ward'].transform(lambda x:x.mode()[0])
df['Ward'].fillna(ward_mode, inplace=True)
```

```python
community_area_mode = df.groupby('District')['Community Area'].transform(lambda x:x.mode()[0])
df['Community Area'].fillna(community_area_mode, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Primary Type'].unique()
```

```python
df = df[(df['Latitude'].between(41.60, 42.10)) & (df['Longitude'].between(-88.0, -87.40))]
```

```python
df['Description'] = df['Description'].str.lower()
df['Location Description'] = df['Location Description'].str.lower()
```

```python
df['Description'] = df['Description'].str.strip()
df['Location Description'] = df['Location Description'].str.strip()
```

```python
df['Description'] = df['Description'].str.replace(r'\s+', ' ', regex=True)
df['Location Description'] = df['Location Description'].str.replace(r'\s+', ' ', regex=True)
```

```python
df['Primary Type'] = df['Primary Type'].str.lower()
```

```python

df.value_counts()
```

```python
df.to_csv('../Data/clean_crime_data.csv', index=False)
```
